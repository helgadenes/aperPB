# Make an HI moment 0 map
This is used to make a map for regridding the compound beam maps

```python
from spectral_cube import SpectralCube
import matplotlib.pyplot as plt
from astropy.wcs import WCS
import astropy.units as u
from astropy.coordinates import FK5
from astropy.coordinates import SkyCoord
from astropy.io import fits
import numpy as np
```

```python
cube = SpectralCube.read('/tank/denes/190409015/B004/HI_image_cube2.fits') 
moment_0 = cube.with_spectral_unit(u.Hz).moment(order=0)

fig = plt.figure(figsize=(18, 12))
ax = fig.add_subplot(111)

# Display the moment map image
im = ax.imshow(moment_0.hdu.data, cmap='Oranges')
```

```python
moment_0.write('/tank/denes/drift_scan/HI_mom0.fits', format='fits') 
```
